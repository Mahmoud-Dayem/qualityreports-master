```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
data = df['SiO2']   # change to your column

fig, ax1 = plt.subplots()

# Normal KDE
sns.kdeplot(data, ax=ax1, label="KDE", fill=True)
ax1.set_ylabel("Density")

# Second axis for cumulative KDE
ax2 = ax1.twinx()

sns.kdeplot(data, ax=ax2, cumulative=True, color="red", label="Cumulative KDE")
ax2.set_ylabel("Cumulative")

# Legends (combine both)
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2)

plt.title("KDE vs Cumulative KDE")
plt.show()
```
